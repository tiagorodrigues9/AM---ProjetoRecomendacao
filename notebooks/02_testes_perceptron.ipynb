# 02. Teste Manual do Perceptron

Neste notebook, vamos importar nosso modelo Perceptron recém-criado e realizar testes manuais para verificar se ele está se comportando como esperado.

```python
import sys
import os
import pandas as pd # Importamos pandas para ler o dataset

# Adiciona o diretório raiz ao path para conseguir importar 'src'
sys.path.append(os.path.abspath('..'))

from src.models.perceptron import Perceptron
```

## 1. Instanciando o Modelo

Vamos criar uma instância do nosso Perceptron. Ele já vem com os pesos padrão que definimos (`energy=0.8`, `loudness=0.2`).

```python
cerebro = Perceptron()
print("Modelo criado! Pesos:", cerebro.weights)
```

## 2. Teste: Música Agitada (Festa)

Vamos simular uma música com alta energia (0.9) e volume alto (-4dB).

```python
resultado_festa = cerebro.predict(energy=0.9, loudness=-4.0)
print(f"Resultado Festa: {resultado_festa}")
# Esperamos prediction=1
```

## 3. Teste: Música Calma (Relax)

Vamos simular uma música com baixa energia (0.2) e volume baixo (-15dB).

```python
resultado_relax = cerebro.predict(energy=0.2, loudness=-15.0)
print(f"Resultado Relax: {resultado_relax}")
# Esperamos prediction=0
```

## 4. Desafio: Testando com Músicas Reais

Vamos carregar o dataset original para buscar músicas reais e ver o que o nosso modelo acha delas.

```python
# Carregando os dados
df = pd.read_csv('../data/raw/dataset.csv')

# Função auxiliar para buscar por ID e prever
def predict_by_id(track_id):
    # Busca a linha
    track = df[df['track_id'] == track_id].iloc[0]
    
    # Extrai features
    name = track['track_name']
    artist = track['artists']
    energy = track['energy']
    loudness = track['loudness']
    
    # Previsão
    result = cerebro.predict(energy, loudness)
    mood = "Festa/Agitada" if result['prediction'] == 1 else "Relax/Calma"
    
    print(f"Música: {name} - {artist}")
    print(f"Dados: Energy={energy}, Loudness={loudness}dB")
    print(f"Previsão do Modelo: {mood} (Ativação: {result['activation']:.4f})")
    print("-" * 30)
```

## 5. Brincando com os Pesos (Tuning)

Você percebeu que algumas músicas com **Energia 1.0** mas **Loudness Baixo** (ex: -16dB) ainda são classificadas como "Festa" (1)?

Isso acontece porque o peso da Energia (0.8) é muito maior que o do Loudness (0.2). Vamos tentar ajustar isso para que o Loudness penalize mais!

```python
# Vamos aumentar o peso do Loudness e diminuir o da Energia
novos_pesos = {'energy': 0.6, 'loudness': 0.5}

# Criando um novo cérebro com esses pesos
cerebro_ajustado = Perceptron(weights=novos_pesos)

# Função de teste rápido
def testar_pesos(modelo, track_id):
    track = df[df['track_id'] == track_id].iloc[0]
    result = modelo.predict(track['energy'], track['loudness'])
    mood = "Festa" if result['prediction'] == 1 else "Relax"
    print(f"Música: {track['track_name']} (Energy={track['energy']}, Loudness={track['loudness']})")
    print(f"Previsão: {mood} (Z = {result['activation']:.4f})")

# Testando com a música problemática (ID: 7FnV73ZJDJL0UklXAd3yfP)
# Exemplo de ID: 7FnV73ZJDJL0UklXAd3yfP (Substitua pelo ID real se diferir)
try:
    #testar_pesos(cerebro_ajustado, "7FnV73ZJDJL0UklXAd3yfP")
    testar_pesos(cerebro_ajustado, "4Xh9F5OQyJjN9Gx7IUiL0n")

except IndexError:
    print("ID não encontrado no dataset. Tente outro ID de exemplo.")
```
