```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import numpy as np
import time
```

```python
# Vetores

# Vetor de entradas : [energy, loudness]
X = [0.9, 0.6]

# Vetor de pesos: [peso_energy, peso_loudness]
W = [0.8, 0.2]

z_manual = (X[0] * W[0]) + (X[1] * W[1])

print(f"Cálculo manual: z = {z_manual}")
```

```python
# Agora com Numpy

X_np = np.array(X)
W_np = np.array(W)

# Produto escalar
z_numpy = np.dot(X_np, W_np)
print(f"Cálculo Numpy: z = {z_numpy}")
print(f"Mesmo resultado ?: {z_manual} == {z_numpy}")
```

### Predição em Lote

```python
# Gerando 10.000 músicas aleatórias
np.random.seed(42)
N = 10_000_000

energy_random = np.random.uniform(0, 1, N)
loudness_random = np.random.uniform(-60, 0, N)

print(f"Geradas {N} músicas aleatórias!")
print(f"Energy:   min={energy_random.min():.2f}, max={energy_random.max():.2f}")
print(f"Loudness: min={loudness_random.min():.2f}, max={loudness_random.max():.2f}")
```

```python
# Teste 1: Loop Python (Perceptron antigo)
inicio = time.time()
for i in range(N):
    modelo_antigo.predict(energy=energy_random[i], loudness=loudness_random[i])
tempo_loop = time.time() - inicio

print(f"Loop Python ({N} músicas): {tempo_loop:.4f}s")
```

```python
# Teste 2: NumPy Batch (Perceptron novo)
X_batch = np.column_stack([energy_random, loudness_random])

inicio = time.time()
modelo_numpy.predict_batch(X_batch)
tempo_numpy = time.time() - inicio

print(f"NumPy Batch ({N} músicas): {tempo_numpy:.4f}s")
```
