# Visualização Simples

```python
plt.figure(figsize=(10,6))
plt.hist(df['danceability'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Danceability')
plt.ylabel('Frequência')
plt.title('Distribuição da dançabilidade')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['energy'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('energy')
plt.ylabel('Frequência')
plt.title('Distribuição da energia')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['energy'], df['loudness'], alpha=0.5)
plt.xlabel('Energia')
plt.ylabel('Barulho')
plt.title('Energia X Barulho')
plt.show()
```

```python
# Filtrando exceções: Alta energia e Baixo Volume
excecoes = df[(df['energy'] > 0.8) & (df['loudness'] < -10)]

# Ordenando pelas mais energéticas
excecoes_ordenadas = excecoes[['track_id', 'track_name', 'artists', 'energy', 'loudness']].sort_values(by='energy', ascending=False)

# Exibindo as top 10
excecoes_ordenadas.head(10)
```
